```python
x = mathdata['N_Terms'][0:50].values
y = mathdata['Total_Cr'][0:50].values

import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.show()
```

```python
m,b = np.polyfit(x,y,1)
#1 being the polynomial degree, in this case just 1 => a line, 2 being a simple curve
plt.plot(x,y,'.')#x,y points, draw it with dots
plt.plot(x,m*x+b,'-')#x, f(x), draw it with a line
plt.show()#show our simple linear regression
```

```python
x = mathdata['N_Terms'][0:50].values
y = mathdata['Total_Cr'][0:50].values

n,m,b = np.polyfit(x,y,2)
lineX = np.linspace(0,max(x),num=100)
lineY = [n*i*i+m*i+b for i in lineX]
plt.plot(x,y,'.')
plt.plot(lineX,lineY,'-')
plt.show()
```

The above is the barbaric method of getting a 2nd degree polynomial least squares fit.

Now to show scikitlearn and how it makes life easier... 

```python
from sklearn.linear_model import LinearRegression

x = mathdata['N_Terms'][0:50].values
y = mathdata['Total_Cr'][0:50].values

model = LinearRegression()

print x,"\n\n"

x = [[i] for i in x]#so yeah .... this is sort of a weird necessity for model.fit in the next line

print x,'\n\n'

y = [[i] for i in y]
model.fit(x,y)
print model.predict([1])
print model.predict([1])[0]#better
```

```python
plt.plot(x,y,'.')
plt.plot(x,model.predict([x])[0],'-')
plt.show()
```

```python
print "r-squared: ",model.score(x,y) #ideally you would use this on test data, not training data
```
