```python
%matplotlib inline
from __future__ import division
import numpy as np
import pandas as pd
import sklearn
from sklearn import metrics
from sklearn.cross_validation import train_test_split, cross_val_score
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

def printClusters( df, cluster_list,k=2):

    categories=['non-graduates','graduates']
    gradClustersDF=pd.DataFrame({'Graduated':df['Graduated'], 'ClusterLbl':cluster_list})
    for i in range(2):
        print 'total',categories[i], len(gradClustersDF[gradClustersDF['Graduated']==i])
        for j in range(k):
            print j,'cluster',\
                gradClustersDF[(gradClustersDF['Graduated']==i)&(gradClustersDF['ClusterLbl']==j)]['ClusterLbl'].count()
    return gradClustersDF

def plotPCA(X_raw,cluster_list,k=2,norm=False,colors='rbgmyck',markers='xDo+v*s.', centered=True):

    pca = PCA(n_components=2)
    #X_raw=df[df.columns[6:]]
    if norm==True:
        normalizer=sklearn.preprocessing.Normalizer()
        X_raw_norm=normalizer.fit_transform(X_raw.transpose().values.astype('float64')).transpose()
    elif centered==True:    
        X_raw_norm= (X_raw - X_raw.mean()) # / (X_raw.max() - X_raw.min())
    else:
        X_raw_norm=X_raw
        
    reduced_X = pca.fit_transform(X_raw_norm)
    
    points=[[[],[]] for j in range(k)]
   
   
    for i in range(len(reduced_X)):
        for j in range(k):
            if cluster_list[i]==j: #red
                points[j][0].append(reduced_X[i][0])
                points[j][1].append(reduced_X[i][1])

    #colors='rbgmyck'   
    #markers='xD.+ov*s'
    for i in range(min(k, len(colors))):
        plt.scatter(points[i][0], points[i][1], c=colors[i],\
                    marker=markers[i], label=str(i)+' Cluster')
        
    plt.legend()
    plt.show()
    
    
def getDistancesDF(columns, cluster_centers,k=2):
    distancesDF=pd.DataFrame({'class':columns,
                          'Cluster0':cluster_centers[0],
                          'Cluster1':cluster_centers[1]})  
    distancesDF=distancesDF[['class','Cluster0','Cluster1']]
    for i in range(2,k):
        distancesDF['Cluster'+str(i)]=cluster_centers[i]
    return distancesDF    
```

```python
import re

def ratePass(x):
    if x[0]!='0':
        if len(x)>3 and x[3]<='C':
            return 1.0/int(x[0])
        else:
            return -1.0/int(x[0])
    else:
        return 0.0

def encode(s):
    v = re.split(r':',s)
    x = v[1][1:]
    out = 0
    if x=='A+':
        out=4.0
    if x=='A':
        out=4.0
    if x=='A-':
        out=3.7
    if x=='B+':
        out=3.3
    if x=='B':
        out=3.0
    if x=='B-':
        out=2.7
    if x=='C+':
        out=2.3
    if x=='C':
        out=2.0
    if x=='C-':
        out=1.7
    if x=='D+':
        out=1.3
    if x=='D':
        out=1.0
    if x=='D-':
        out=0.7
    if x=='F':
        out=0.0
    if x=='NA':
        out = 2.0
    if x=='CR':
        out=2.0
    #return [int(v[0]),out-2]
    #return int(v[1][0])*out-1-int(v[0])
    return int(v[1][0])*out-2 #incorporate credit data
```

```python
ratedDF=data[data.columns[5:]].applymap(ratePass)
ratedDF[data.columns[:5]]=data[data.columns[:5]]
ratedDF=ratedDF[data.columns]

gpaDF=data[data.columns[5:]].applymap(encode)
gpaDF[data.columns[:5]]=data[data.columns[:5]]
gpaDF=gpaDF[data.columns]

gpaDF=data[data.columns[5:]].applymap(encode)
gpaDF[data.columns[:5]]=data[data.columns[:5]]
gpaDF=gpaDF[data.columns]
```

```python
from sklearn.cluster import KMeans
k=2
cl=5

#center=clusterCenters(data, k=k,n_tests=10,init='random')
kmeans_model = KMeans(n_clusters=k, n_init=20).fit(gpaDF[gpaDF.columns[cl:]])
gradClustersDF=printClusters(ratedDF, kmeans_model.labels_,k=kmeans_model.n_clusters)
plotPCA(ratedDF[gpaDF.columns[cl:]],kmeans_model.labels_,k=kmeans_model.n_clusters,norm=False, centered=False)

confusion_matrix = metrics.confusion_matrix(gradClustersDF['Graduated'], gradClustersDF['ClusterLbl'])
print "confustion matrix numbers"
print(confusion_matrix)
plt.matshow(confusion_matrix)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
```

```python
false_positive_rate_tr, recall_tr, thresholds_tr = metrics.roc_curve(gradClustersDF['Graduated'],gradClustersDF['ClusterLbl'])

roc_auc_tr = metrics.auc(false_positive_rate_tr, recall_tr)
plt.title('Receiver Operating Characteristic')
plt.plot(false_positive_rate_tr, recall_tr, 'b', label='AUC = %0.2f' %roc_auc_tr)
plt.legend(loc='lower right')
plt.plot([0, 1], [0, 1], 'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.ylabel('Recall')
plt.xlabel('Fall-out')
plt.show()

print metrics.classification_report(gradClustersDF['Graduated'],gradClustersDF['ClusterLbl'])

#distancesDF=getDistancesDF(gpaDF.columns[cl:], kmeans_model.cluster_centers_,k=kmeans_model.n_clusters)
#distancesDF['Distance']=abs(distancesDF['Cluster0']-distancesDF['Cluster1'])
#distancesDF.sort('Distance', ascending=False)
```

```python
#applying regression to Graduated vs Total_Cr
from sklearn.linear_model import LogisticRegression
import sklearn
from sklearn import metrics
from sklearn.cross_validation import train_test_split, cross_val_score
import matplotlib.pyplot as plt


col = [5,6,7,8,10,14]
X_train, X_test, y_train, y_test = train_test_split(gpaDF[gpaDF.columns[col]], data['Graduated'])
'''
X_train=data['Total_Cr'].values
X_test=data['Total_Cr'].values
y_train=data['Graduated'].values
y_test=data['Graduated'].values
'''

classifier = LogisticRegression()
classifier.fit(X_train.reshape(len(X_train),len(col)), y_train)
y_pred = classifier.predict(X_test.reshape(len(X_test),len(col)))

confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
print "confustion matrix numbers"
print(confusion_matrix)
plt.matshow(confusion_matrix)
plt.title('Confusion matrix')
plt.colorbar()
plt.ylabel('True label')
plt.xlabel('Predicted label')
classifier.coef_
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
#classifier.score(X_test,y_test) #same metric
scores = cross_val_score(classifier, X_train, y_train, cv=5)
print 'Accuracy:  ', np.mean(scores), scores

precisions = cross_val_score(classifier, X_train, y_train, cv=5, scoring='precision')
print 'Precision: ', np.mean(precisions), precisions

recalls = cross_val_score(classifier, X_train, y_train, cv=5, scoring='recall')
print 'Recall:    ', np.mean(recalls), recalls

f1s = cross_val_score(classifier, X_train, y_train, cv=5, scoring='f1')
print 'F1: ', np.mean(f1s), f1s
```

```python
from sklearn.metrics import roc_curve, auc

predictions = classifier.predict_proba(X_test)
false_positive_rate, recall, thresholds = roc_curve(y_test,predictions[:, 1])
roc_auc = auc(false_positive_rate, recall)
plt.title('Receiver Operating Characteristic')
plt.plot(false_positive_rate, recall, 'b', label='AUC = %0.2f' %roc_auc)
plt.legend(loc='lower right')
plt.plot([0, 1], [0, 1], 'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.ylabel('Recall')
plt.xlabel('Fall-out')
plt.show()
```
